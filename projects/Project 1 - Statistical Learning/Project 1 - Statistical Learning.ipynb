#### Importing necessary libraries

```python
import numpy as  np
import pandas as pd
```

Let's replace dashes(-) with 0

```python
data_df.replace(to_replace='-', value=0, inplace=True)
data_df.sample(8)
```

We have many int column as object. We need change their dtype

```python
data_df = data_df.astype({
    'Seasons': 'int64', 
    'Points': 'int64',
    'GamesPlayed': 'int64', 
    'GamesWon': 'int64',
    'GamesDrawn': 'int64',
    'GamesLost': 'int64', 
    'GoalsFor': 'int64', 
    'GoalsAgainst': 'int64', 
    'Champion': 'int64',
    'Runner-up': 'int64', 
    'Third': 'int64', 
    'Fourth': 'int64', 
    'Fifth': 'int64', 
    'Sixth': 'int64', 
    'T': 'int64'
})
data_df.dtypes
```

```python
data_df.columns
```

#### Q. What is winning percent for each team?

```python
data_df['WinningPercent'] = np.round((data_df['GamesWon']/data_df['GamesPlayed']) * 100, 2)
```

```python
data_df.nlargest(5, 'WinningPercent')[['Team', 'WinningPercent', 'Pos']]
```

#### Q. What is Debut year range for all teams?
We are tryinng to determine whether all team are old and stable or there are still new teams coming?

```python
data_df['Debut_year'] = data_df['Debut'].astype(str).str[:4].astype('int64')

print(f"First team debut is in the year {data_df['Debut_year'].min()} \nLatest Team debut is in th year {data_df['Debut_year'].max()}")
```

#### Which team debuted from 1930 to 1980 and what is there current position?
Here we trying to see which has been playing from the start and get the glimpse of which are more stable ones

```python
team_debut_in_50_years = data_df[data_df['Debut'].astype(str).str[:4].astype('int64').between(1929,1980)][['Team', 'Pos', 'Debut', 'Debut_year']].sort_values('Debut')
team_debut_in_50_years
```

#### Q. What is the Goal Difference for each Team i.e. GoalFor - Goal Against?
We are trying too see which team have scored more goals in favor than goal against them. This will help us understand which teams are more strong than other teams

```python
data_df['GoalDifference'] = data_df.apply(lambda x: (x['GoalsFor'] - x['GoalsAgainst']), 1)
data_df[['Team', 'Pos', 'GoalDifference']].head()
```

#### Q. Which have highest goal difference in favour and against? (Top 5 )

```python
data_df.nlargest(5, 'GoalDifference')[['Team', 'Pos']]
```

```python
data_df.nsmallest(5, 'GoalDifference')[['Team', 'Pos']]
```

#### Q. What is the mean average winning percent and Goal difference for teams finising on best positions?

```python
group_by_best_position = data_df.groupby('BestPosition')
```

```python
# By Mean
group_by_best_position[['Team', 'BestPosition', 'Pos', 'WinningPercent', 'GoalDifference']].mean()#.head().sort_values(by='BestPosition')
```

```python
# By Median
group_by_best_position[['Team', 'BestPosition', 'WinningPercent', 'GoalDifference']].median()
```

Let's see the teams position sorted by best position

```python
data_df.sort_values(by='BestPosition')[['Team', 'BestPosition', 'Pos', 'WinningPercent', 'GoalDifference']].head(15)
```
