## Dataset Information

The data at hand contains medical costs of people characterized by certain attributes.
Information regarding the dataset:

- **age**: age of primary beneficiary
- **sex**: insurance contractor gender, female, male
- **bmi**: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9
- **children**: Number of children covered by health insurance / Number of dependents
- **smoker**: Smoking
- **region**: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.
- **charges**: Individual medical costs billed by health insurance.

#### Q. Do charges of people who smoke differ significantly from the people who don't?
Let have following hypothesis:
- Null hypothesis: charges of people who smoke do not differ significantly from the people who don't. Mu = 0
- Alternative : charges of people who smoke differ significantly from the people who don't. Mu > 0

```python
smoker_condition = insurance_df['smoker'] == 1
```

```python
# Let's check mean and variance for both the samples
print('For non smoker \nMean = %3.2f, SD = %3.2f, var = %3.2f' %(np.mean(insurance_df[~smoker_condition]['charges']), 
                                                                 np.std(insurance_df[~smoker_condition]['charges']), 
                                                                 np.var(insurance_df[~smoker_condition]['charges'])))
print('For smoker \nMean = %3.2f, SD = %3.2f, var = %3.2f' %(np.mean(insurance_df[smoker_condition]['charges']), 
                                                              np.std(insurance_df[smoker_condition]['charges']), 
                                                              np.var(insurance_df[smoker_condition]['charges'])))
```

```python
# Let's try the problem with 2 sample test
tstatistic, pval =stats.ttest_ind(insurance_df[~smoker_condition]['charges'], 
                                  insurance_df[smoker_condition]['charges'], 
                                  equal_var='False')

print("p-value = ", pval)

if(pval < 0.05):
    print("\nAs per the 2 sample test, Null hypothesis is rejected i.e. charges of people who smoke differ significantly from the people who don't.")
else:
    print("\nAs per the 2 sample test, we Failed to reject Null Hypothesis i.e. charges of people who smoke do not differ significantly from the people who don't.")
```

```python
# Let's try to use ANOVA (one sided)
# It's recommended to use ANOVA if we have more than two sample.But it work on the two sample as well.

tstatistic, pval =stats.f_oneway(insurance_df[~smoker_condition]['charges'], 
                                 insurance_df[smoker_condition]['charges'])

print("p-value = ", pval)

if(pval < 0.05):
  print("\nAs per the ANOVA test, Null hypothesis is rejected i.e. charges of people who smoke differ significantly from the people who don't.")
else:
  print("\nAs per the ANOVA test, we Failed to reject Null Hypothesis i.e. charges of people who smoke do not differ significantly from the people who don't.")
```

This is same we visualized using plots.

#### Q. Does bmi of males differ significantly from that of females?

Let have following hypothesis:

- Null hypothesis: bmi of males do not differ significantly from that of females. Mu = 0
- Alternative : bmi of males differ significantly from that of females. Mu > 0

```python
sex_condition = insurance_df['sex'] == 0
```

```python
# two sample test 
tstatistic, pval =stats.ttest_ind(insurance_df[sex_condition]['bmi'], 
                                  insurance_df[~sex_condition]['bmi'])

print("p-value = ", pval)

if(pval < 0.05):
    print("\nAs per the 2 sample test, Null hypothesis is rejected i.e. bmi of males differ significantly from that of females.")
else:
    print("\nAs per the 2 sample test, Failed to reject Null Hypothesis i.e. bmi of males do not differ significantly from that of females")
```

This is same we visualized using plots.

#### Q. Is the distribution of bmi across women with no children, one child and two children, the same ?

- Null hypothesis: distribution of bmi across women with no children, one child and two children is same
- Alternate hypothesis: distribution of bmi across women with no children, one child and two children is not same

```python
# bmi of women with 0 children
women_bmi_zero_child = insurance_df[(insurance_df['sex']==0) & (insurance_df['children'] == 0)]['bmi']

# bmi of women with 1 children
women_bmi_one_child = insurance_df[(insurance_df['sex']==0) & (insurance_df['children'] == 1)]['bmi']

# bmi of women with 0 children
women_bmi_two_child = insurance_df[(insurance_df['sex']==0) & (insurance_df['children'] == 2)]['bmi']

# since we have three sample using one way ANOVA
tstatistic, pval =stats.f_oneway(women_bmi_zero_child, women_bmi_one_child, women_bmi_two_child)
print("statics", tstatistic)
print("p-value = ", pval)

if(pval < 0.05):
    print("\n AS per the ANOVA test, Null hypothesis is rejected i.e. distribution of bmi across women with no children, one child and two children is not same.")
else:
    print("\n AS per the ANOVA test, Failed to reject Null Hypothesis i.e. distribution of bmi across women with no children, one child and two children is same.")
```
